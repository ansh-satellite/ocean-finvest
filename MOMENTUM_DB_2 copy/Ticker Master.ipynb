```python
#!/usr/bin/env python
# coding: utf-8

import os
import pandas as pd
import numpy as np
import yfinance as yf

# -------------------------------
# 🔹 Folder path
# -------------------------------
folder_path = r'Universe'   # ⬅️ Change this to your folder path

# -------------------------------
# 🔹 Initialize list to hold DataFrames
# -------------------------------
all_dfs = []

# -------------------------------
# 🔹 Loop through all files in folder
# -------------------------------
for file in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file)
    
    # Skip subfolders
    if not os.path.isfile(file_path):
        continue
    
    try:
        if file.lower().endswith('.csv'):
            df = pd.read_csv(file_path)
        elif file.lower().endswith(('.xlsx', '.xls')):
            df = pd.read_excel(file_path)
        else:
            continue  # skip non-CSV/XLS files
        
        df['SourceFile'] = file
        all_dfs.append(df)
    except Exception as e:
        print(f"⚠️ Error reading {file}: {e}")

# -------------------------------
# 🔹 Combine all DataFrames
# -------------------------------
if all_dfs:
    combined_df = pd.concat(all_dfs, ignore_index=True, sort=False)
    print(f"✅ Successfully combined {len(all_dfs)} files!")

    # -------------------------------
    # 🔹 Add Ticker column (assuming 'Symbol' exists)
    # -------------------------------
    if 'Symbol' in combined_df.columns:
        combined_df['Ticker'] = combined_df['Symbol'].astype(str) + ".NS"
    else:
        print("⚠️ 'Symbol' column not found. Skipping ticker creation.")

    # -------------------------------
    # 🔹 Fetch Sector and Industry Info
    # -------------------------------
    unique_tickers = np.sort(combined_df['Ticker'].dropna().unique())
    info_mapping = {}

    for ticker in unique_tickers:
        print(f"Fetching info for {ticker}...")
        try:
            info = yf.Ticker(ticker).info
            info_mapping[ticker] = {
                'Sector': info.get('sector', 'N/A'),
                'Industry': info.get('industry', 'N/A'),
                'Market Cap': info.get("marketCap", 'N/A'),
                'Beta': info.get("beta", 'N/A')      
            }
        except Exception as e:
            print(f"⚠️ Error fetching {ticker}: {e}")
            info_mapping[ticker] = {'Sector': 'N/A', 'Industry': 'N/A'}

    # Convert mapping to DataFrame
    info_df = pd.DataFrame.from_dict(info_mapping, orient='index').reset_index()
    info_df = info_df.rename(columns={'index': 'Ticker'})
info_df
```
